# Tensor Crash Course


```python
import torch
import numpy as np
```

```python
v = torch.tensor([1,2,3,4,5,6])
print('dtype:', v.dtype)
print('shape:', v.shape)
print('item 1:', v[0])

print('Slice tensor:', v[1:4])


```

```python
f = torch.FloatTensor([1,2,3,4,5,6,7])
print('dtype:', f.dtype)
print('size:', f.size())
print('item 1:', f[0])

print('Slice tensor:', f[1:4])


```

```python
v.view(3,-1)
```

```python
a = np.array([1,2,3,4,5,6])
tensor_converted = torch.from_numpy(a)
print(tensor_converted)



```
