```python
import torch
import matplotlib.pyplot as plt
```

## Operaciones con tensores

```python
t_one = torch.tensor([1,2,3,4,5,6])
t_two = torch.tensor([1,2,3,4,5,6])

sum = t_one + t_two
mul = t_one * t_two
product = t_one * 5
dot_product = torch.dot(t_one, t_two)

#--------------------------------

print('sum:', sum)
print('mul:', mul)
print('product:', product)
print('dot_product:', dot_product)
```

## Funciones utiles de tensores

```python
x = torch.linspace(0,10,10)
y = torch.sigmoid(x)
# y = torch.exp(x)

print(x)
print(y)
```
